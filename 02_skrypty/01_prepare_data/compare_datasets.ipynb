```python
import pandas as pd
```

```python
path = "C:\\Users\\julian\Desktop\studia\economic analytics\Projekt Badawczy - BIK\\01_dane\\01_dane_raw\\"

bik_1 = pd.read_csv(f"{path}BIK 1.csv", delimiter=";")
bik_2 = pd.read_csv(f"{path}BIK 2.csv", delimiter=";")
bik_3 = pd.read_csv(f"{path}BIK 3.csv", delimiter=";")
bik = pd.read_csv(f"{path}BIK.csv", delimiter=";")
bik2 = pd.read_csv(f"{path}BIK2.csv", delimiter=";")
proba_bik2010 = pd.read_csv(f"{path}proba_BIK201012.csv", delimiter=";")
```

```python
dataset_list = [bik_1, bik_2, bik_3, bik, bik2, proba_bik2010]
dataset_names = ["bik_1", "bik_2", "bik_3", "bik", "bik2", "proba_bik2010"]

for dataset, name in zip(dataset_list, dataset_names):
    dataset["dataset_name"] = name
```

```python
all_datasets = pd.concat(dataset_list, ignore_index=True)
```

```python
all_datasets.groupby("dataset_name", as_index=False).describe(percentiles=[]).to_excel(
    "describe_summary.xlsx"
)
```
